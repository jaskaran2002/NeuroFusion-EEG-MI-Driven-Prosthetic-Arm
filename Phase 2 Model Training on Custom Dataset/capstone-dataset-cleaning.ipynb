```python
import numpy as np
import pandas as pd
```

```python
import os
```

```python
FILENAMES = [i for i in os.listdir('/kaggle/input/capstone-dataset/Capstone Dataset') if '.csv' in i and '_intervalMarker.csv' not in i and 'rest,' not in i]
FILENAMES = sorted(FILENAMES)
FILENAMES
```

```python
for FILENAME in FILENAMES:
    # FILENAME = 'full-trial-4_EPOCFLEX_252029_2024.12.04T17.21.12+05.30.md.bp.csv'
    # DELETE_LINE_NUMBER = 1
    
    with open(f'/kaggle/input/capstone-dataset/Capstone Dataset/{FILENAME}') as f:
        data = f.read().splitlines() # Read csv file
    data = data[1:]
    for i in range(len(data)):
        data[i] = ','.join(data[i].split(',')[:16])
    with open(f'/kaggle/working/{FILENAME}', 'w') as g:
        g.write('\n'.join(data)) # Write to file
    df = pd.read_csv(f'/kaggle/working/{FILENAME}')
    # print(df.columns)
    print(df.shape)
```

```python
!mkdir /kaggle/working/data
```

```python
for filename in FILENAMES:
    if 'hand-trial-1' in filename:
        continue
    df = pd.read_csv(f'/kaggle/working/{FILENAME}')
    df = df[['EEG.Cz', 'EEG.C3', 'EEG.FC5', 'EEG.T7', 'EEG.CP5', 'EEG.O1', 'EEG.Pz',
       'EEG.O2', 'EEG.CP6', 'EEG.T8', 'EEG.FC6', 'EEG.C4']]
    df['target'] = -1
    name = filename.split('_')[0]
    if ('arm' in filename):
        df.loc[128*60: 2*128*60, 'target'] = 0
        df.loc[2*128*60: 3*128*60, 'target'] = 4
        df.loc[3*128*60: , 'target'] = 3
    elif ('full' in filename):
        # full
        df.loc[128*60: 1.5*128*60, 'target'] = 0
        df.loc[1.5*128*60: 2*128*60, 'target'] = 3
        df.loc[2*128*60: 2.5*128*60, 'target'] = 1
        df.loc[2.5*128*60: 3*128*60, 'target'] = 2
        df.loc[3*128*60: , 'target'] = 4
    elif ('hand' in filename):
        # hand
        df.loc[128*60: 2*128*60, 'target'] = 0
        df.loc[2*128*60: 3*128*60, 'target'] = 2
        df.loc[3*128*60: , 'target'] = 1
    df.to_csv(f'/kaggle/working/data/{name}.csv')
```
